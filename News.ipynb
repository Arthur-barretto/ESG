### Scraping

```python
import urllib.request,sys,time
from bs4 import BeautifulSoup
import requests
```

```python
dic = {}
pgs = list(range(2,20))
pgs.insert(0, '')

for i in pgs:
    url = 'https://www.politifact.com/factchecks/list/?page='+str(i)
    print(url)
    try:
        page=requests.get(url) 
        dic[i] = page
    except Exception as e:    
        error_type, error_obj, error_info = sys.exc_info()      

        #print the link that cause the problem
        print ('ERROR FOR LINK:',url)

        #print error info and line that threw the exception                          
        print (error_type, 'Line:', error_info.tb_lineno)

        continue
```

```python
for pg in pgs:
    soup = BeautifulSoup(dic[pg].text, "html.parser")
    
    
    source = {}
    cont = 0
    for div in soup.findAll('div', {'class': 'm-statement__meta'}):
        source[cont] = div.find('a').text.strip()
        cont += 1
        
    stated = {}
    cont = 0
    for div in soup.findAll('div', {'class': 'm-statement__meta'}):
        stated[cont] = div.find('div', {'class': 'm-statement__desc'}).text.strip()
        cont += 1
    
    quotes = {}
    cont = 0
    for div in soup.findAll('div', {'class': 'm-statement__quote'}):
        quotes[cont] = div.find('a').text.strip()
        cont += 1
        
    upload_by = {}
    cont = 0
    for footer in soup.findAll('footer', {'class': 'm-statement__footer'}):
        upload_by[cont] = footer.text.strip()
        cont += 1    
        
    d = {'source':source.values(), 'stated': stated.values(), 'quotes':quotes.values(), 'upload_by':upload_by.values()}
    df = pd.DataFrame(data=d)
```
